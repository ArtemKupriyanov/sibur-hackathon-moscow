```python
plains_repare = pd.read_csv("Вагоны и плановые ремонты.csv")
```

```python
plains_repare[:4]
```

```python
plain_reps = ['std_kap', 'std_ppr', 'std_dep', 'std_vogi']
start_date = pd.to_datetime("2018-05-01")
end_date = pd.to_datetime("2018-06-30")
```

```python
idxs_all = np.zeros(len(plains_repare))
for col in plain_reps:
    plains_repare[col] = pd.to_datetime(plains_repare[col])
    idx = (start_date <= plains_repare[col]).values * (plains_repare[col] <= end_date).values
    idxs_all = idxs_all + idx
```

Капитальный
Планово-предупредительный
Деповский
Вакуумная очистка и гидроиспытания

```python
plains_repare_may_june = plains_repare[idxs_all > 0]
```
